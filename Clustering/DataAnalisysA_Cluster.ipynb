```python
cycleA = pd.read_csv("../Data/Chickens data/WithMacro/CycleA-Clean.csv")
```

```python
cycleA["Eggs"].sum()
```

```python
parameters = ["DeathPerc" , "Laied", "AvgTemp", "AvgHum"]

meansA = []
for p in parameters:
    if p == "DeathPerc":
        meansA.append(cycleA[p].mean()*100)
    else:
        meansA.append(cycleA[p].mean())


means = [meansA]

ranges = []

for i in range(len(parameters)):
    min = means[0][i]
    max = means[0][i]
    for m in means:
        if m[i] < min:
            min = m[i]
        if m[i] > max:
            max = m[i]
    ranges.append([min - min/4, max + max/11])
```
