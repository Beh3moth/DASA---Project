```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import sklearn
import scipy
np.random.seed(0)
```

```python
cycleA_1 = pd.read_csv("Data/Chickens data/CycleA-1.csv", sep=";")
cycleA_2 = pd.read_csv("Data/Chickens data/CycleA-2.csv", sep=";")
```

```python
cycleA_1.head()
```

```python
cycleA_2.head()
```

```python
cycleA_1 = cycleA_1[cycleA_1["Chickens Death Per Day"]<100]
```

```python
# input (x) and output (y) features
y_features = ["Chickens Death Per Day", "Current Chickens", "# of Eggs"]

# pair plot cycle A
#norm_a = np.array()
sns.set(style="ticks", color_codes=True)
sns.pairplot(cycleA_1[y_features])
plt.show()

# heatmap cycle A
sns.heatmap(cycleA_1[y_features].corr(), annot=True)
plt.show()
# found a strong correlation between:
#   -   FEED and EGGS
#   -   WATER and DEATH

# pair plot cycle B
sns.set(style="ticks", color_codes=True)
sns.pairplot(cycleA_2[y_features])
plt.show()

# heatmap cycle B
sns.heatmap(cycleA_2[y_features].corr(), annot=True)
plt.show()
```
