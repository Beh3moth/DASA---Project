```python
prices = pd.read_csv("../Data/EggsData/price_comparison.csv")
```

```python
def TemporalCluster(dataSet):

    fig, ax1 = plt.subplots()
    ax1.set_xlabel('Date')
    ax1.set_ylabel("â‚¬/100pz")
    #Select to plot just years
    ax1.xaxis.set_major_locator(mpl.dates.YearLocator())
    ax1.xaxis.set_minor_locator(mpl.dates.MonthLocator())
    ax1.xaxis.set_major_formatter(mpl.dates.DateFormatter("\n%Y"))
    ax1.xaxis.set_minor_formatter(mpl.dates.DateFormatter("%b"))
    fig.set_figwidth(20)
    plt.setp(ax1.get_xticklabels(), rotation=0, ha="center")

    ax1.plot(dataSet["Date"],dataSet["Non-Organic"], label="Non-Organic")
    ax1.plot(dataSet["Date"],dataSet["Organic"], label="Organic")
    ax1.legend()


    plt.gcf().autofmt_xdate()
    plt.xticks(rotation=45)
    plt.show()
```

```python
prices["Date"] = pd.to_datetime(prices["Date"], format = '%d/%m/%Y')
```

```python
TemporalCluster(prices)
```
